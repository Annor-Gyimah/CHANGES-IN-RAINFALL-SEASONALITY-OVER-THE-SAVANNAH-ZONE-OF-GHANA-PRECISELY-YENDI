```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv('Yendi.csv',header=0)
mapping={Data.columns[0]: 'Years'}
Data.rename(columns=mapping, inplace=True)
Data.set_index('Years',inplace=True)
del Data['1']
del Data['2']
del Data['5']
del Data['7']
del Data['8']
del Data['10']
del Data['12']
#mapping={Data.columns[0]: 'Years',Data.columns[1]:'Onset-Days', Data.columns[2]:'Cessation-Days', Data.columns[3]: 'Length-of-Gro-Season', Data.columns[4]:'Intensity'}
mapping={Data.columns[0]: 'Onset-Days',Data.columns[1]:'Cessation-Days', Data.columns[2]:'Length-of-Gro-Season', Data.columns[3]: 'Intensity'}
Data.rename(columns=mapping, inplace=True)
#Data.set_index('Years', inplace=True)
Data.to_csv('YENDII.csv')
Data
```

```python
Data = pd.read_csv('YENDII.csv',usecols=['Years','Onset-Days','Cessation-Days','Length-of-Gro-Season','Intensity'], sep =',')
Onset = Data.loc[:,'Onset-Days']
Onset_m = np.mean(Onset)
Cessation = Data.loc[:,'Cessation-Days']
Cessation_m = np.mean(Cessation)
LOGS = Data.loc[:,'Length-of-Gro-Season']
LOGS_m = np.mean(LOGS)
Intensity = Data.loc[:,'Intensity']
Intensity_m = np.mean(Intensity)
Years = Data.loc[:,'Years']
```

```python

fig, ax = plt.subplots(figsize = (20, 20))
fig.subplots_adjust(bottom=0.5, left = 0.4)
# On-Set Days Plot
ax.plot(Years,Onset, color = 'k', lw = 3, marker = 'o', label = 'On-Set Days')
ax.set_ylabel('On-Set Days', fontsize =18)
ax.set_xlabel('Years', fontsize =18)
ax.set_title('On-Set Days Plot', size =20, weight ='bold')
ax.axhline(y = Onset_m, color ='r', label = 'Mean On-Set', linestyle = '--', lw =3)
ax.set_xticks(range(1960, 2020,5))
ax.set_yticks(range(70, 150, 5))
ax.legend(loc = 2, fontsize =15)
ax.legend(title = 'Legend',loc =1, bbox_to_anchor=(0.25, 1.0), fontsize = 15)

```

```python
fig, ax = plt.subplots(figsize = (20, 20))
fig.subplots_adjust(bottom=0.5, left = 0.4)
# Cessation Days Plot
ax.plot(Years,Cessation, color = 'b', lw = 3, marker = 'o', label = 'Cessation Days')
ax.set_ylabel('Cessation Days', fontsize =18)
ax.set_xlabel('Years', fontsize =18)
ax.set_title('Cessation Days Plot', size =20, weight ='bold')
ax.axhline(y=Cessation_m, color ='r', label = 'Mean Cessation Days', linestyle = '--', lw =3)
ax.set_xticks(range(1960, 2020, 5))
ax.legend(loc = 3, fontsize =15)
ax.legend(title = 'Legend', loc=2, bbox_to_anchor=(0.65, 1), fontsize = 15)


```

```python
fig, ax = plt.subplots(figsize = (20, 20))
fig.subplots_adjust(bottom=0.5, left = 0.4)
# Lenght of Growing Season Plot
ax.plot(Years,LOGS, color = 'g', lw = 3, marker = 'o', label = 'Length of Growing Season')
ax.set_ylabel('Length of Growing Season', fontsize =18)
ax.set_xlabel('Years', fontsize =18)
ax.set_title('Length of Growing Season Plot', size =20, weight ='bold')
ax.axhline(y = LOGS_m, color ='r', label = 'Mean Length of Growing Season', linestyle = '--', lw =3)
ax.set_xticks(range(1960, 2020, 5))
ax.legend(title = 'Legend', loc=2, bbox_to_anchor=(-0.01, 0.17), fontsize = 15)
```

```python
fig, ax = plt.subplots(figsize = (20, 20))
fig.subplots_adjust(bottom=0.5, left = 0.4)
# Intensity Plot
ax.plot(Years,Intensity, color = 'y', lw = 3, marker = 'o', label = 'Intensity')
ax.set_ylabel('Intensity', fontsize =18)
ax.set_xlabel('Years', fontsize =18)
ax.set_title('Intensity Plot', size =20, weight ='bold')
ax.axhline(y = Intensity_m, color ='r', label = 'Mean Intensity', linestyle = '--', lw =3)
ax.set_xticks(range(1960, 2020, 5))
ax.set_yticks(range(7, 21, 1))
ax.legend(title = 'Legend',loc = 2, fontsize =15)
```
