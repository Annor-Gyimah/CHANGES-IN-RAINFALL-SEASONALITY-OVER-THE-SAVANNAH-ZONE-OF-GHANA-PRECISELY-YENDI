```python
df = pd.read_csv(r'dl.csv',usecols=[1,2,3])
mapping = {df.columns[0]: 'Month',df.columns[1]:'Day',df.columns[2]:'Rainfall'}
df.rename(columns=mapping, inplace=True)
#df.to_csv('dl.csv')
rf = df.loc[:,'Rainfall'].sum()
bf = df.loc[:,'Rainfall']
bf = (bf/rf)*100
ff = bf.cumsum()

bf['Perca'] = bf
ff['CumRain'] = ff
# #rf = rf.sum()
df['Perca'] = bf['Perca']
df['CumRain'] = ff['CumRain']
```
